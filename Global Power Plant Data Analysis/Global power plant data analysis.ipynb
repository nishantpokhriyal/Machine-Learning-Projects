## Content

The Global Power Plant Database is a comprehensive, open source database of power plants around the world. It centralizes power plant data to make it easier to navigate, compare and draw insights for one’s own analysis. The database covers approximately 35,000 power plants from 167 countries and includes thermal plants (e.g. coal, gas, oil, nuclear, biomass, waste, geothermal) and renewables (e.g. hydro, wind, solar). Each power plant is geolocated and entries contain information on plant capacity, generation, ownership, and fuel type. It will be continuously updated as data becomes available.
But our dataset contains data for India only.

### Taking care of Nan values

**We are going to fill the nan values with mean and mode of the particular column**

**We will replace missing values of longitude and latitude with the most frequent values of the columns and missing values of generation_gwh with the mean of their columns.**

```python
# Applying Simple Imputer to convert the nan values into numeric
from sklearn.impute import SimpleImputer

imp_mean=SimpleImputer(missing_values=np.nan,strategy='mean')
imp_mode=SimpleImputer(missing_values=np.nan,strategy='most_frequent')

pp['latitude']=imp_mode.fit_transform(pp['latitude'].values.reshape(-1,1))
pp['longitude']=imp_mode.fit_transform(pp['longitude'].values.reshape(-1,1))
pp['generation_gwh_2014']=imp_mean.fit_transform(pp['generation_gwh_2014'].values.reshape(-1,1))
pp['generation_gwh_2015']=imp_mean.fit_transform(pp['generation_gwh_2015'].values.reshape(-1,1))
pp['generation_gwh_2016']=imp_mean.fit_transform(pp['generation_gwh_2016'].values.reshape(-1,1))
pp['generation_gwh_2017']=imp_mean.fit_transform(pp['generation_gwh_2017'].values.reshape(-1,1))
pp['generation_gwh_2018']=imp_mean.fit_transform(pp['generation_gwh_2018'].values.reshape(-1,1))
```

```python
# Creating Variance Inflation Factor function
pp_=pp.copy()
pp_.drop(columns=['name','other_fuel1','primary_fuel'],inplace=True)         # They are not numeric

from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

ppc_=add_constant(pp_)
def ppc_vif():
    vif=pd.DataFrame()
    vif['Variables']=ppc_.columns
    vif['VIF Factor']=[variance_inflation_factor(ppc_.values,i)
                       for i in range(ppc_.shape[1])]
    return (vif)
```

## Removing Outliers

```python
pp.head()
```

```python
# Checking presence of skewness in the data
pp.skew()
```

**High Skewness is present in all the variables except latitude**

**Now we will convert the categorical variable "other_fuel1" into numeric.**

**We will apply Label Encoding to variable "other_fuel1".**

```python
# Converting elements of other_fuel1 into numeric
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()

pp['other_fuel1']=l.fit_transform(pp['other_fuel1'])
```

```python
# Removing Outliers
from numpy import percentile

a=pp.columns
for i in a:
    if isinstance(pp[i][0],str):
        continue
    else:
        # Quartiles
        q=percentile(pp[i],[25,75])
        # Lower & Upper bound
        lower_bound= q[0] - (1.5*(q[1]-q[0]))
        upper_bound= q[1] + (1.5*(q[1]-q[0]))
        pp[i] = pp[i].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
```

**We have removed the outliers higher than the upper whisker by the value of the upper whisker and the outliers lower than the lower whisker by the value of the lower whisker.**

```python
from sklearn.model_selection import train_test_split
xf_train,xf_test,yf_train,yf_test=train_test_split(xf,yf,test_size=0.20,random_state=50)
```

**We've splitted our dataset into training and testing, where we have used our 80% of data for training and 20% of data for testing**

### Scaling x_train and x_test using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

xf_train=sc.fit_transform(xf_train)
xf_test=sc.transform(xf_test)
```

```python
# Importing important algorithms and metrics to create the prediction models and to check their accuracy
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

**Making a function to calculate and print following features for the model : Accuracy Score, Confusion Matrix, Classification Report.**

```python
def classifier_func(y_test,pred):
    ac=accuracy_score(y_test,pred)
    cm=confusion_matrix(y_test,pred)
    cr=classification_report(y_test,pred)
    
    print('Accuracy Score :',ac,'\n')
    print('Confusion Matrix :',cm,'\n')
    print('Classification Report :',cm,'\n')
```

**Making a Cross Validation function**

```python
from sklearn.model_selection import cross_val_score
def cross_val_accuracy(model,xf_train=xf_train,yf_train=yf_train):
    accuracies=cross_val_score(estimator=model,X = xf_train, y = yf_train ,cv=10)
    print('Cross_val_Accuracy: {:.2f} %'.format(accuracies.mean()*100),'\n')
    print('Cross_val_Standard Deviation: {:.2f} %'.format(accuracies.std()*100),'\n')
```

### Decision Tree Classifier

**Creating Decision Tree Classifier Model, it'll fit the Decision Tree Classifier model on xf_train and yf_train and predict the xf_test and store it in pred_dtc. We will print predicted Primary Fuel and actual Primary Fuel and outputs of the above functions**

```python
dtc=DecisionTreeClassifier()
dtc.fit(xf_train,yf_train)
pred_dtc=dtc.predict(xf_test)

print('\033[1m' + 'DECISION TREE CLASSIFIER MODEL SUMMERY\n' + '\033[0m')
classifier_func(yf_test,pred_dtc)
cross_val_accuracy(dtc)

print('Predicted Primary Fuel',pred_dtc)
print('\nActual Primary Fuel\n',yf_test)
```

### KNeighbors Classifier

**Creating KNeighbors Classifier Model, it'll fit the KNeighbors Classifier model on xf_train and yf_train and predict the xf_test and store it in pred_knc. We will print predicted Primary Fuel and actual Primary Fuel and outputs of the above functions**

```python
knc=KNeighborsClassifier()
knc.fit(xf_train,yf_train)
pred_knc=knc.predict(xf_test)

print('\033[1m' + 'KNEIGHBORS CLASSIFIER MODEL SUMMERY\n' + '\033[0m')
classifier_func(yf_test,pred_knc)
cross_val_accuracy(knc)

print('Predicted Primary Fuel',pred_knc)
print('\nActual Primary Fuel\n',yf_test)
```

### Random Forest Classifier

**Creating Random Forest Classifier Model, it'll fit the Random Forest Classifier model on xf_train and yf_train and predict the xf_test and store it in pred_rc. We will print predicted Primary Fuel and actual Primary Fuel and outputs of the above functions**

```python
rc=RandomForestClassifier(n_estimators=100,random_state=42)
rc.fit(xf_train,yf_train)
pred_rc=rc.predict(xf_test)

print('\033[1m' + 'RANDOM FOREST CLASSIFIER MODEL SUMMERY\n' + '\033[0m')
classifier_func(yf_test,pred_rc)
cross_val_accuracy(rc)

print('Predicted Primary Fuel',pred_rc)
print('\nActual Primary Fuel\n',yf_test)
```

## Model Creation

```python
pp.head()
```

```python
x=pp.iloc[:,1:]
y=pp.iloc[:,0]

print(x.shape)
print(y.shape)
```

**We have created two variables containing Input data (x) and Output data (y).**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=50)
```

**We've splitted our dataset into training and testing, where we have used our 80% of data for training and 20% of data for testing**

### Scaling x_train and x_test using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# Importing linear Regression algorithm, mean absolute error, mean squared error and r2 score to make the prediction model and checking its accuracy
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
```

**Making a function to calculate and print following features for the model : mean absolute error, mean squared error, r2 score.**

```python
def results(actual,pred):
    mae=mean_absolute_error(actual,pred)
    mse=mean_squared_error(actual,pred)
    r2=r2_score(actual,pred)
    
    print('Mean Absolute Error :',mae,'\n')
    print('Mean Squared Error :',mse,'\n')
    print('r2 Score :',r2,'\n')
```

**Making a Cross Validation function**

```python
from sklearn.model_selection import cross_val_score
```

```python
def accuracy(model,x_train=x_train,y_train=y_train):
    accuracies=cross_val_score(estimator=model,X = x_train, y = y_train ,cv=10)
    print('Cross_val_Accuracy: {:.2f} %'.format(accuracies.mean()*100),'\n')
    print('Cross_val_Standard Deviation: {:.2f} %'.format(accuracies.std()*100),'\n')
```

### Linear Regression Model

**Creating Linear Regression Model, it'll fit the regression model on x_train and y_train and predict the x_test and store it in pred. We will print predicted capacity_mw and actual capacity_mw and outputs of the above functions**

```python
ln=LinearRegression()
ln.fit(x_train,y_train)
pred=ln.predict(x_test)

print('\033[1m' + 'LINEAR REGRESSION MODEL SUMMERY\n' + '\033[0m')
print('Predicted Capacity_mw',pred)
print('\nActual Capacity_mw\n',y_test)
print('\n')
results(y_test,pred)
accuracy(ln)
```

### Lasso Regression

**Creating Lasso Regression Model, it'll fit the regression model on x_train and y_train and predict the x_test and store it in pred_ls. We will print predicted capacity_mw and actual capacity_mw and outputs of the above functions**

```python
from sklearn.linear_model import Lasso
```

```python
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
pred_ls=ls.predict(x_test)

print('\033[1m' + 'LASSO REGRESSION MODEL SUMMERY\n' + '\033[0m')
print('Predicted Capacity_mw',pred_ls)
print('\nActual Capacity_mw\n',y_test)
print('\n')
results(y_test,pred_ls)
accuracy(ls)
```

## Gradient Boosting Regressor

**Creating Linear Regression Model, it'll fit the regression model on x_train and y_train and predict the x_test and store it in pred_gb. We will print predicted capacity_mw and actual capacity_mw and outputs of the above functions**

```python
from sklearn.ensemble import GradientBoostingRegressor
gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
pred_gb=gb.predict(x_test)

print('\033[1m' + 'GRADIENT BOOSTING REGRESSOR MODEL SUMMERY\n' + '\033[0m')
print('Predicted Capacity_mw',pred_gb)
print('\nActual Capacity_mw\n',y_test)
print('\n')
results(y_test,pred_gb)
accuracy(gb)
```
