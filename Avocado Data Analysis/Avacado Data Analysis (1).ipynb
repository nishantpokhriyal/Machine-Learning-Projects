## Removing Outliers

```python
av.head()
```

```python
av.skew()
```

```python
from numpy import percentile

a=av.columns
for i in a:
    if isinstance(av[i][0],str):
        continue
    else:
        # Quartiles
        q=percentile(av[i],[25,75])
        # Lower & Upper bound
        lower_bound= q[0] - (1.5*(q[1]-q[0]))
        upper_bound= q[1] + (1.5*(q[1]-q[0]))
        av[i] = av[i].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
```

**We have removed the outliers higher than the upper whisker by the value of the upper whisker and the outliers lower than the lower whisker by the value of the lower whisker.**

## Model Creation

```python
avr.head()
```

```python
xr=avr.drop(columns=['region'])
yr=avr[['region']]
print(xr.shape)
print(yr.shape)
```

**We have created two variables containing Input data (xr) and Output data (yr).**

```python
from sklearn.model_selection import train_test_split
xr_train,xr_test,yr_train,yr_test=train_test_split(xr,yr,test_size=0.20,random_state=50)
```

**We've splitted our dataset into training and testing, where we have used our 80% of data for training and 20% of data for testing**

### Scaling x_train and x_test using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
xr_train=sc.fit_transform(xr_train)
xr_test=sc.transform(xr_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

**Making a function to calculate and print following features for the model : Accuracy Score, Confusion Matrix, Classification Report.**

```python
def classifier_func(yr_test,pred_r):
    ac=accuracy_score(yr_test,pred_r)
    cm=confusion_matrix(yr_test,pred_r)
    cr=classification_report(yr_test,pred_r)
    
    print('Accuracy Score :',ac,'\n')
    print('Confusion Matrix :',cm,'\n')
    print('Classification Report :',cm,'\n')
```

**Making a Cross Validation function for "Region"**

```python
from sklearn.model_selection import cross_val_score
def r_accuracy(model,xr_train=xr_train,yr_train=yr_train):
    accuracies=cross_val_score(estimator=model,X = xr_train, y = yr_train ,cv=10)
    print('Cross_val_Accuracy: {:.2f} %'.format(accuracies.mean()*100),'\n')
    print('Cross_val_Standard Deviation: {:.2f} %'.format(accuracies.std()*100),'\n')
```

### Decision Tree Classifier

**Creating Decision Tree Classifier Model, it'll fit the Decision Tree Classifier model on xr_train and yr_train and predict the xr_test and store it in pred_dtc. We will print predicted average price and actual average price and outputs of the above functions**

```python
dtc=DecisionTreeClassifier()
dtc.fit(xr_train,yr_train)
pred_dtc=dtc.predict(xr_test)

print('\033[1m' + 'DECISION TREE CLASSIFIER MODEL SUMMERY\n' + '\033[0m')
classifier_func(yr_test,pred_dtc)
r_accuracy(dtc)
```

### KNeighbors Classifier

**Creating KNeighbors Classifier Model, it'll fit the KNeighbors Classifier model on xr_train and yr_train and predict the xr_test and store it in pred_knc. We will print predicted average price and actual average price and outputs of the above functions**

```python
knc=KNeighborsClassifier()
knc.fit(xr_train,yr_train)
pred_knc=knc.predict(xr_test)

print('\033[1m' + 'KNEIGHBORS CLASSIFIER MODEL SUMMERY\n' + '\033[0m')
classifier_func(yr_test,pred_knc)
r_accuracy(knc)
```

### Support Vector Classifier

**Creating Support Vector Classifier Model, it'll fit the Support Vector Classifier model on xr_train and yr_train and predict the xr_test and store it in pred_svc. We will print predicted average price and actual average price and outputs of the above functions**

```python
svc=SVC()
svc.fit(xr_train,yr_train)
pred_svc=svc.predict(xr_test)

print('\033[1m' + 'SUPPORT VECTOR CLASSIFIER MODEL SUMMERY\n' + '\033[0m')
classifier_func(yr_test,pred_svc)
r_accuracy(svc)
```

## Model Creation

```python
av.head()
```

```python
x=av.iloc[:,1:]
y=av.iloc[:,0]

print(x.shape)
print(y.shape)
```

**We have created two variables containing Input data (x) and Output data (y).**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=50)
```

**We've splitted our dataset into training and testing, where we have used our 80% of data for training and 20% of data for testing**

### Scaling x_train and x_test using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
```

**Making a function to calculate and print following features for the model : mean absolute error, mean squared error, r2 score.**

```python
def results(actual,pred):
    mae=mean_absolute_error(actual,pred)
    mse=mean_squared_error(actual,pred)
    r2=r2_score(actual,pred)
    
    print('Mean Absolute Error :',mae,'\n')
    print('Mean Squared Error :',mse,'\n')
    print('r2 Score :',r2,'\n')
```

**Making a Cross Validation function**

```python
from sklearn.model_selection import cross_val_score
```

```python
def accuracy(model,x_train=x_train,y_train=y_train):
    accuracies=cross_val_score(estimator=model,X = x_train, y = y_train ,cv=10)
    print('Cross_val_Accuracy: {:.2f} %'.format(accuracies.mean()*100),'\n')
    print('Cross_val_Standard Deviation: {:.2f} %'.format(accuracies.std()*100),'\n')
```

### Linear Regression Model

**Creating Linear Regression Model, it'll fit the regression model on x_train and y_train and predict the x_test and store it in pred. We will print predicted average price and actual average price and outputs of the above functions**

```python
ln=LinearRegression()
ln.fit(x_train,y_train)
pred=ln.predict(x_test)

print('\033[1m' + 'LINEAR REGRESSION MODEL SUMMERY\n' + '\033[0m')
print('Predicted Average Price',pred)
print('\nActual Average Price\n',y_test)
print('\n')
results(y_test,pred)
accuracy(ln)
```

```python
from yellowbrick.regressor import ResidualsPlot
```

```python
plt.figure(figsize=(17,10),dpi=100)
plot=ResidualsPlot(ln,hist=True)
plot.score(x_test,y_test)
plot.show()
```

```python
rr=RandomForestRegressor()
rr.fit(x_train, y_train)
pred_rr=rr.predict(x_test)

print('\033[1m' + 'RANDOM FOREST REGRESSOR MODEL SUMMERY\n' + '\033[0m')
print('Predicted Average Price',pred_rr)
print('\nActual Average Price\n',y_test)
print('\n')
results(y_test,pred_rr)
accuracy(rr)
```

```python
plt.figure(figsize=(17,10),dpi=100)
plot=ResidualsPlot(rr,hist=True)
plot.score(x_test,y_test)
plot.show()
```
