# Removing Outliers

```python
fb.head()
```

```python
from numpy import percentile
      
a = fb.columns

for i in a:
    if pd.api.types.is_numeric_dtype(fb[i]):
        q = percentile(fb[i], [25, 75])
        lower_bound = q[0] - (1.5 * (q[1] - q[0]))
        upper_bound = q[1] + (1.5 * (q[1] - q[0]))
        fb[i] = np.where(fb[i] < lower_bound, lower_bound, np.where(fb[i] > upper_bound, upper_bound, fb[i]))
```

**we have identified and removed the outliers from the numeric columns in the dataset 'fb'. The code checks if a column is numeric, calculates the quartiles and determines the lower and upper bounds. Then, it replaces any value below the lower bound with the lower bound value and any value above the upper bound with the upper bound value. This helps to remove extreme values that could skew the analysis and improve the accuracy of the results.**
